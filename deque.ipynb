### 나만의 deque 자료구조 만들기

#### Python list 관련 함수만을 활용할 것

#### mydeque = MyDeque(): 덱 생성

#### mydeque.func("push_front X"): 정수 X를 덱의 앞에 넣는다.
#### mydeque.func("push_back X"): 정수 X를 덱의 뒤에 넣는다.
#### mydeque.func("pop_front"): 덱의 가장 앞에 있는 수를 빼고, 그 수를 출력한다. 만약, 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
#### mydeque.func("pop_back"): 덱의 가장 뒤에 있는 수를 빼고, 그 수를 출력한다. 만약, 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
#### mydeque.func("size"): 덱에 들어있는 정수의 개수를 출력한다.
#### mydeque.func("empty"): 덱이 비어있으면 1을, 아니면 0을 출력한다.
#### mydeque.func("front"): 덱의 가장 앞에 있는 정수를 출력한다. 만약 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
#### mydeque.func("back"): 덱의 가장 뒤에 있는 정수를 출력한다. 만약 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.

#### print(mydeque): 현재 덱 출력
#### mydeque: 현재 덱 출력
#### 현재 덱 출력 결과: e.g. 덱에 들어있는 값이 1, 2, 3 정수들일때 --> mydeque([1, 2, 3])

#### 본인이 만든 deque 모든 함수들의 사용예시 각자 제시할 것

```python
class mydeque:
    def __init__(self):
        self.myd = []
        self.amt = 0
    def __str__(self):
        return "mydeque([{}])".format(",".join(self.myd))
    def __repr__(self):
        return "mydeque([{}])".format(",".join(self.myd))
    def func(self,order):
        order = order.split(' ')
        if "push_front" in order:
            self.myd.insert(0,order[-1])
        elif "push_back" in order:
            self.myd.append(order[-1])
        elif order[0] == "pop_front":
            if len(self.myd)==0:
                return -1
            else:
                return self.myd.pop(0)
        elif order[0] == "pop_back":
            if len(self.myd) == 0:
                return -1
            else:
                return self.myd.pop()
        elif order[0] is "size":
            return len(self.myd)
        elif order[0] is "empty":
            if len(self.myd) == 0:
                return 1
            else:
                return 0
        elif order[0] is "front":
            if len(self.myd)== 0:
                return -1
            else:
                return self.myd[0]
        elif order[0] is "back":
            if len(self.myd) == 0:
                return -1
            else:
                return self.myd[len(self.myd)-1]
        else:
            print("그런건없음")
```

```python
md = mydeque()
md.func("push_front 3")
print(md)
md.func("push_back 4")
print(md)
md.func("push_front 5")
print(md)
print(md.func("size"))
print(md.func("front"))
print(md.func("back"))
print(md.func("pop_front"))
print(md.func("pop_back"))
print(md)
print(md.func("size"))
print(md.func("empty"))
print(md.func("pop_front"))
print(md)
print(md.func("empty"))
print(md.func("으아악!"))
```
